```python
import numpy as np
import matplotlib.pyplot as plt

# Simulated values for beta and alpha, and corresponding model performance
beta_values = np.linspace(0.1, 1.0, 10)
performance_beta = -0.5 * (beta_values - 0.5)**2 + 0.95  # Parabolic performance curve for beta

alpha_values = np.linspace(0.1, 2.0, 10)
performance_alpha = -0.1 * (alpha_values - 1.0)**2 + 0.92  # Parabolic performance curve for alpha

# Create the plot for beta sensitivity
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.plot(beta_values, performance_beta, marker='o', linestyle='-', color='blue')
plt.title("Impact of $\\beta$ on Model Performance")
plt.xlabel("$\\beta$")
plt.ylabel("Model Performance")
plt.grid(True)

# Create the plot for alpha sensitivity
plt.subplot(1, 2, 2)
plt.plot(alpha_values, performance_alpha, marker='o', linestyle='-', color='green')
plt.title("Impact of $\\alpha$ on Model Performance")
plt.xlabel("$\\alpha$")
plt.ylabel("Model Performance")
plt.grid(True)

# Show the plots
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Beta values and corresponding R2 scores for the PPB task
beta_values = [0, 1e-5, 1e-4, 1e-3, 1e-1]
r2_scores = [0.636, 0.6455, 0.6498, 0.6462, 0.5992]

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(beta_values, r2_scores, marker='o', linestyle='-', color='b')

# Set the scale of x-axis to logarithmic
plt.xscale('log')

# Add labels and title
plt.xlabel(r'beta Value (log scale)')
plt.ylabel('R^2 Score for PPB')
plt.title('Sensitivity Analysis: R^2 Score vs. beta Value for PPB Task')
plt.grid(True)

# Display the plot
plt.show()
```
